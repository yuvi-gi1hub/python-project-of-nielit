## ðŸŽ¯ Objective of the Project

The objective of this project is to perform Exploratory Data Analysis (EDA) on a COVID-19 dataset using Python libraries such as NumPy, Pandas, Matplotlib, and Seaborn.  
We aim to visualize and understand the impact of COVID-19 by analyzing trends in confirmed cases, deaths, and recoveries across different countries.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="darkgrid")
```

```python
# Load the COVID-19 dataset
url = 'https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv'
df = pd.read_csv(url)

# Show the first 5 rows
df.head()
```

```python
# Dataset Information
df.info()
```

```python
# Statistical Summary
df.describe()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Check for missing values
df.isnull().sum()
```

```python
# Histogram of Confirmed Cases
plt.figure(figsize=(10,6))
plt.hist(df['Confirmed'], bins=50, color='skyblue')
plt.title('Distribution of Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart of Total Confirmed Cases by Country (Top 10)
country_confirmed = df.groupby('Country')['Confirmed'].max().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=country_confirmed.index, y=country_confirmed.values, palette='viridis')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xlabel('Country')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Confirmed vs Deaths
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Confirmed', y='Deaths', hue='Country', legend=False)
plt.title('Confirmed vs Deaths')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.show()
```

```python
# Heatmap of Correlation
plt.figure(figsize=(8,5))
sns.heatmap(df[['Confirmed', 'Recovered', 'Deaths']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Confirmed, Recovered, and Deaths')
plt.show()
```

## Time series analysis

### Subtask:
Analyze trends of confirmed cases, deaths, and recoveries over time globally.


**Reasoning**:
Group the DataFrame by 'Date', calculate the sum of 'Confirmed', 'Recovered', and 'Deaths', and then plot the trends over time for each category.



```python
global_trends = df.groupby('Date')[['Confirmed', 'Recovered', 'Deaths']].sum().reset_index()

plt.figure(figsize=(15, 7))
plt.plot(global_trends['Date'], global_trends['Confirmed'], label='Confirmed', color='blue')
plt.plot(global_trends['Date'], global_trends['Recovered'], label='Recovered', color='green')
plt.plot(global_trends['Date'], global_trends['Deaths'], label='Deaths', color='red')

plt.title('Global COVID-19 Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

## Country-specific analysis

### Subtask:
Analyze and visualize the trends for specific countries of interest.


**Reasoning**:
Select a few countries of interest and filter the dataframe to include data only for these countries.



```python
countries_of_interest = ['United States', 'India', 'Brazil', 'United Kingdom']
filtered_df = df[df['Country'].isin(countries_of_interest)].copy()
```

**Reasoning**:
Group the filtered data by 'Date' and 'Country' and calculate the sum of 'Confirmed', 'Recovered', and 'Deaths' for each group, then reset the index.



```python
country_trends = filtered_df.groupby(['Date', 'Country'])[['Confirmed', 'Recovered', 'Deaths']].sum().reset_index()
```

**Reasoning**:
Create a multi-line plot to visualize the trends of confirmed cases, recovered cases, and deaths for each of the selected countries over time, using different colors or line styles to distinguish between countries and case types, adding appropriate titles, labels, and a legend.



```python
plt.figure(figsize=(15, 8))

sns.lineplot(data=country_trends, x='Date', y='Confirmed', hue='Country', linestyle='-', label='Confirmed', errorbar=None)
sns.lineplot(data=country_trends, x='Date', y='Recovered', hue='Country', linestyle='--', label='Recovered', errorbar=None)
sns.lineplot(data=country_trends, x='Date', y='Deaths', hue='Country', linestyle=':', label='Deaths', errorbar=None)

plt.title('COVID-19 Trends for Selected Countries Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend(title='Case Type and Country', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The error indicates that the `label` argument is being passed multiple times to the `lineplot` function when using `hue`. When `hue` is used, seaborn automatically handles the legend based on the `hue` variable. To fix this, remove the `label` argument from the `sns.lineplot` calls and handle the legend manually to show both country and case type.



```python
plt.figure(figsize=(15, 8))

sns.lineplot(data=country_trends, x='Date', y='Confirmed', hue='Country', linestyle='-', errorbar=None)
sns.lineplot(data=country_trends, x='Date', y='Recovered', hue='Country', linestyle='--', errorbar=None)
sns.lineplot(data=country_trends, x='Date', y='Deaths', hue='Country', linestyle=':', errorbar=None)

plt.title('COVID-19 Trends for Selected Countries Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')

# Create custom legend handles
handles, labels = plt.gca().get_legend_handles_labels()
case_type_handles = [plt.Line2D([], [], color='gray', linestyle='-', label='Confirmed'),
                     plt.Line2D([], [], color='gray', linestyle='--', label='Recovered'),
                     plt.Line2D([], [], color='gray', linestyle=':', label='Deaths')]

# Combine country and case type legends
plt.legend(handles=handles + case_type_handles, labels=labels + ['Confirmed', 'Recovered', 'Deaths'],
           title='Country and Case Type', loc='upper left', bbox_to_anchor=(1, 1))

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Comparison analysis

### Subtask:
Compare the trends across different countries.


**Reasoning**:
Group the data by country to find the maximum confirmed, recovered, and death counts for each country and sort by confirmed cases to get the top 10.



```python
latest_country_data = df.groupby('Country')[['Confirmed', 'Recovered', 'Deaths']].max().reset_index()
top_10_countries_confirmed = latest_country_data.sort_values(by='Confirmed', ascending=False).head(10)
```

**Reasoning**:
Create a bar plot to visualize the total confirmed cases for the top 10 countries.



```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Confirmed', data=top_10_countries_confirmed, palette='viridis')
plt.title('Top 10 Countries by Total Confirmed Cases')
plt.xlabel('Country')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

**Reasoning**:
Sort the data by recovered cases and create a bar plot for the top 10 countries.



```python
top_10_countries_recovered = latest_country_data.sort_values(by='Recovered', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Recovered', data=top_10_countries_recovered, palette='viridis')
plt.title('Top 10 Countries by Total Recovered Cases')
plt.xlabel('Country')
plt.ylabel('Total Recovered Cases')
plt.xticks(rotation=45)
plt.show()
```

**Reasoning**:
Sort the data by death cases and create a bar plot for the top 10 countries to complete the comparison of trends across different countries.



```python
top_10_countries_deaths = latest_country_data.sort_values(by='Deaths', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Deaths', data=top_10_countries_deaths, palette='viridis')
plt.title('Top 10 Countries by Total Deaths')
plt.xlabel('Country')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.show()
```

## Correlation analysis

### Subtask:
Explore the correlation between confirmed cases, deaths, and recoveries using appropriate visualizations.


**Reasoning**:
Calculate the correlation matrix and create a heatmap to visualize the correlation between 'Confirmed', 'Recovered', and 'Deaths' columns.



```python
# Calculate the correlation matrix
correlation_matrix = df[['Confirmed', 'Recovered', 'Deaths']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of COVID-19 Cases')
plt.show()
```
